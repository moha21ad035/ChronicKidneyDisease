```python
# object means it predicting the datatype incorrectly
# Whenever Pandas does not recognize the data type as one of the small handful of datatypes it can deal with (int, float, string, boolean, …), 
#it just sets the datatype to “object” 
```

```python
# this how we object datatypes to correct datatype which the actual datatype that variable is caring

def convert_text_to_numeric(df,column):
    df[column] = pd.to_numeric(df[column],errors="coerce")
for column in text_columns:
    convert_text_to_numeric(df,column)
    print(f"{column} : {df[column].dtype}")
    
```

```python
#handling missing values null values
#whenever we are using square brackets we should consider them as a data frame and then we are sorting them in decending
#sort_values is a syntax for sorting inside the parameter is must whether its a ascending or decending
missing = df.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
```

```python
# how can we fill values in that missing place means we can fill mean,mode,mediun but cant use this for all column
#fillna() is use in column we are specifying itll fill null value intead of we are giving values
def mean_value_imputation(df,column):
    mean_value = df[column].mean()
    df[column].fillna(value=mean_value,inplace=True) # mean is number only(Average)
def mode_value_imputation(df,column):
    mode_value = df[column].mode()[0] #-> in a column there will be lots of mode values we are just taking first.mode means category if we gave mode it only 
    # give us majority vlaue which one is majority in that column
    df[column] = df[column].fillna(mode_value)   
```

```python
# this how we can handle value when our output is numerical by mean
num_cols = [col for col in df.columns if df[col].dtype != "object"] # -> not object which numerical column we are mentioning here
for columns in num_cols:
    mean_value_imputation(df,columns)
```

```python
# this how we can handle value when our output is categorical by mode
num_cols = [col for col in df.columns if df[col].dtype == "object"] #-> object which means not number values columns 
for columns in num_cols:
    mode_value_imputation(df,columns)
```

```python
# replacing values
df["Diabetis_mellitus"] = df["Diabetis_mellitus"].replace(to_replace= {' yes':'yes','\tno':'no','\tyes':'yes'})
df["coronary_artery_disease"] = df["coronary_artery_disease"].replace(to_replace= {' yes':'yes','\tno':'no','\tyes':'yes'})
df["Class"] = df["Class"].replace(to_replace={"ckd\t":"ckd","notckd":"not ckd"}) 
```

```python
print(f"Diabetis_mellitus : {df['Diabetis_mellitus'].unique()}") 
print(f"coronary_artery_disease : {df['coronary_artery_disease'].unique()}") 
print(f"Class: {df['Class'].unique()}")
```

```python
# we are going to do mapping because our model doesnt understrand string value only 0 or 1 so we are going convert that
#how are we going to do that using mapping beacause here only two category if its not means we have to other methods
df['Class'] = df["Class"].map({"ckd":1,"not ckd" : 0})
df['Red_blood_cell'] = df["Red_blood_cell"].map({"normal":1,"abnormal" : 0})
df['Pus_cell'] = df["Pus_cell"].map({"normal":1,"abnormal" : 0})
df['Pus_cell_clumps'] = df["Pus_cell_clumps"].map({"present":1,"notpresent" : 0})
df['Bacteria'] = df["Bacteria"].map({"present":1,"notpresent" : 0})
df['Hypertension'] = df["Hypertension"].map({"yes":1,"no" : 0})
df['Diabetis_mellitus'] = df["Diabetis_mellitus"].map({"yes":1,"no" : 0})
df['coronary_artery_disease'] = df["coronary_artery_disease"].map({"yes":1,"no" : 0})
df['appetite'] = df["appetite"].map({"good":1,"poor" : 0})
df['peda_edema'] = df["peda_edema"].map({"yes":1,"no" : 0})
df['Anemia'] = df["Anemia"].map({"yes":1,"no" : 0})
```

```python
from sklearn.model_selection import train_test_split
```

```python
# deleting the Class because we are going to predict the class value (target value) using the feartures existing feature without target value
x = df.drop(["Class"],axis=1)
y = df["Class"]
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.25,random_state=25)

print(f"X shape: {X_train.shape}")
print(f"Xtest shape: {X_test.shape}")
```

```python
from sklearn.tree import DecisionTreeClassifier

dct = DecisionTreeClassifier()
dct.fit(X_train,Y_train)
```

```python
y_pred_dct = dct.predict(X_test)
y_pred_dct
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
models = []
models.append(("Naive Bayes",GaussianNB()))
models.append(("KNN",KNeighborsClassifier(n_neighbors=8)))
models.append(("RandomForestClassifier",RandomForestClassifier()))
models.append(("DecisionTreeClassifier",DecisionTreeClassifier()))
models.append(("SVM",SVC(kernel="linear")))
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
for name,model in models:
    print(name,model)
    print()
    model.fit(X_train,Y_train)
    y_pred = model.predict(X_test)
    print(confusion_matrix(Y_test,y_pred))
    print("\n")
    print(accuracy_score(Y_test,y_pred))
    print("\n")
    print(precision_score(Y_test,y_pred))
    print("\n")
    print(recall_score(Y_test,y_pred))
    print("\n")
    print(f1_score(Y_test,y_pred))
    print("\n")
```
